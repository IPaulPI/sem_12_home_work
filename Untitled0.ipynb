```python
import pandas as pd
import random
from tabulate import tabulate

# Генерация исходного DataFrame
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

# Преобразование в one-hot представление
data_one_hot = pd.DataFrame()

# Создание столбцов для каждого уникального значения в столбце 'whoAmI'
for unique_value in data['whoAmI'].unique():
    data_one_hot[unique_value] = data['whoAmI'].apply(lambda x: 1 if x == unique_value else 0)

# Вывод всей таблицы на экран с использованием tabulate
print(tabulate(data_one_hot, headers='keys', tablefmt='grid'))
```
